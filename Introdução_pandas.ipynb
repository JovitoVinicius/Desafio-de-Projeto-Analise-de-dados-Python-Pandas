# Python para análise de dados (Pandas)

```python
#Importando biblioteca Pandas
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/ESTUDOS/Cursos/Ciência de dados - Bootcamp DIO/datasets/Gapminder.csv", sep=";")
```

```python
#Visualizando as 5 primeiras linhas, através do método head(). Que por padão retorna sempre 5 valores caso não seja expecificado detro do paramêtro.
df.head()
```

```python
#Renomeando as colunas através do método rename que tem como paramêtro um dicionário onde a chave é o valor que vai ser alterado e o valor será o que irá substituir. Lembre-se de usar esse comando entre "{}".
df = df.rename(columns={"country":"País","continent":"continente","year":"Ano","lifeExp":"Expectativa_Vida", "pop":"Pop_Total", "gdpPercap": "PIB"})
```

```python
#Retornando a quantidade de linhas e colunas do df através do método shape
df.shape
```

```python
#Retornando os nomes das colunas do df através do método columns
df.columns
```

```python
#Retornando o tipo de dado de cada coluna do df através do método dtypes
df.dtypes
```

```python
#Retornando as ultimas 5 linhas do df através do método tail(), que por padrão retorna 5 valores caso não seja expecificado um valor dentro do paramêtro.
df.tail()
```

```python
#Retornando dados estatisticos do df através do método describe()
df.describe()
```

```python
#Filtrando dados do df para achar os países presentes no dataframe, através do método unique()
df["continente"].unique()
```

```python
#Filtrando dados do df para localizar os paises de determinado continente, através do método loc
Asia = df.loc[df["continente"] == "Asia"]
Asia.head(10)
```

```python
Asia.shape
```

```python
#Realizando a contagem de paises. Através do método nunique()
df.groupby("continente")["País"].nunique()
```

```python
#Realizando a expectativa de vida média pra cada pais. Através do método mean()
df.groupby("Ano")["Expectativa_Vida"].mean()
```
