```python
#Criando um DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Receita total?
df["Valor Venda"].sum()
```

```python
#Custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando uma nova coluna
```

```python
df.head(1)
```

```python
#Custo total?
round(df["Custo"].sum(), 2)
```

```python
#Com a receita total e o custo total, pode-se encontrar o lucro total.
df["Lucro Total"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
#Total Lucro
round(df["Lucro Total"].sum(),2)
```

```python
#Criando uma tabela de tempo de envio
df["Tempo de Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head()
```

```python
#Transformando a coluna Tempo de Envio para valores inteiros
df["Tempo de Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#Média de tempo de envio para cada marca
df.groupby(df["Marca"])["Tempo de Envio"].mean()
```

```python
#Verificando se há dados faltantes
df.isnull().sum()
```

```python
#Lucro por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum()
```

```python
#Comando para retirar a notação ciêntifica.
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Armazendando a consulta abaixo em uma nova tabela, através do método reset_index() após a atribuição.
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum().reset_index()
```

```python
#Total de Produtos Vendidos
df.groupby(df["Produto"])["Quantidade"].sum().sort_values(ascending=True)
```

```python
#Gráfico Total de produtos vendidos
df.groupby(df["Produto"])["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Total Produtos Vendidos", color = "green")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Lucro por ano 
df.groupby(df["Data Venda"].dt.year)["Lucro Total"].sum().plot.bar(title = "Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro Total"].sum()
```

```python
#Analizando venda do ano de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Lucro por Mês 
df.groupby(df["Data Venda"].dt.month)["Lucro Total"].sum().plot(title = "Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Lucro por Marca
df.groupby(df["Marca"])["Lucro Total"].sum().plot.bar(title = "Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```

```python
#Lucro por Classe
df.groupby(df["Classe"])["Lucro Total"].sum().plot.barh(title = "Lucro x Classe")
plt.xlabel("Lucro")
plt.ylabel("Classe");
```
